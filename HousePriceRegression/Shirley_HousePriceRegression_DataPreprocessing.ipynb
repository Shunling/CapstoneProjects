## Preprocessing with Dummy Table 

```python
df = pd.read_csv('./data/df_clean.csv',index_col='Id')
df['SoldTime'] = pd.to_datetime(df['SoldTime'])
df.head(2)
```

```python
df.info()
```

```python
i = 0
for col in df.columns:
    if df[col].dtypes == 'object':
        df[col] = df[col].astype('category')
        i += 1

df.MSSubClass = df.MSSubClass.astype('category')
i += 1

cat_cols = df.columns[:i]
num_cols = df.columns[i:]
```

```python
y = df['SalePrice']
dummies = df[num_cols].drop('SalePrice',axis=1)
```

```python
for col in cat_cols:
    dummy_col = pd.get_dummies(df[[col]], prefix_sep = '_')

    dummies = dummies.join(dummy_col, on='Id')
```

```python
dummies.shape
```

```python
dummies.columns
```

```python
X_train, X_test, y_train, y_test = train_test_split(dummies, y, test_size=0.3, random_state=7)
```

```python
X_train.head(2)
```

```python
# Save all processed data into folder

X_train.to_csv('./dummy_data/X_train.csv')
X_test.to_csv('./dummy_data/X_test.csv')
y_train.to_csv('./dummy_data/y_train.csv')
y_test.to_csv('./dummy_data/y_test.csv')
```
